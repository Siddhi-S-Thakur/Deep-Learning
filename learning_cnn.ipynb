#### In deep learning, Precision measures the accuracy of positive predictions (how many predicted positives are truly positive), while Recall (Sensitivity) measures how many actual positives the model captured (True Positives / All Actual Positives)

#### If a model finds 8 dogs from 12 total, with 5 true dogs (True Positives) and 3 cats (False Positives), its Precision is 5/8 (62.5%), while if 7 actual dogs were missed (False Negatives), its Recall is 5/12 (41.7%).
